```python
import matplotlib.pyplot as plt
from pathlib import Path
import os
import csv

base_dir = Path(os.getcwd()).parent

history = []

with open(base_dir / 'logs' / 'run_with_cosine_scheduler' /'history.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        history.append(row)
```

```python
import pandas as pd

df = pd.DataFrame(history[1:], columns=history[0])
df = df.astype(float)
```

```python
plt.figure(figsize=(10, 7))

# RMSE Plot
plt.subplot(3, 1, 1)
plt.plot(df.index, df['train_rmse'], label='Train RMSE')
plt.plot(df.index, df['val_rmse'], label='Validation RMSE')
plt.ylabel('RMSE')
plt.legend()

# R^2 Plot
plt.subplot(3, 1, 2)
plt.plot(df.index, df['train_r2'], label='Train R^2')
plt.plot(df.index, df['val_r2'], label='Validation R^2')
plt.xlabel('Epochs')
plt.ylabel('R^2')
plt.legend()

# LR
plt.subplot(3, 1, 3)
plt.plot(df.index, df['lr'], label='learning rate')
plt.xlabel('Epochs')
plt.ylabel('lr')
plt.legend()

plt.tight_layout()
plt.show()
```
