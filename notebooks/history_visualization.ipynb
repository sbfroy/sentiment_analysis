```python
import matplotlib.pyplot as plt
from pathlib import Path
import os
import csv

base_dir = Path(os.getcwd()).parent

history_1, history_2, history_3 = [], [], []

with open(base_dir / 'logs' / 'run_20250106_101731' /'history.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        history_1.append(row)

with open(base_dir / 'logs' / 'run_20250106_111311' /'history.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        history_2.append(row)

with open(base_dir / 'logs' / 'run_20250106_121101' /'history.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        history_3.append(row)
```

```python
import pandas as pd

df_1 = pd.DataFrame(history_1[1:], columns=history_1[0])
df_1 = df_1.astype(float)

df_2 = pd.DataFrame(history_2[1:], columns=history_2[0])
df_2 = df_2.astype(float)

df_3 = pd.DataFrame(history_3[1:], columns=history_3[0])
df_3 = df_3.astype(float)
```

```python
plt.figure(figsize=(10, 7))

# RMSE Plot
plt.subplot(3, 1, 1)
plt.plot(df_1.index, df_1['train_rmse'], label='Train RMSE')
plt.plot(df_1.index, df_1['val_rmse'], label='Validation RMSE')
plt.ylabel('RMSE')
plt.legend()

# R^2 Plot
plt.subplot(3, 1, 2)
plt.plot(df_1.index, df_1['train_r2'], label='Train R^2')
plt.plot(df_1.index, df_1['val_r2'], label='Validation R^2')
plt.xlabel('Epochs')
plt.ylabel('R^2')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 7))

# RMSE Plot
plt.subplot(3, 1, 1)
plt.plot(df_2.index, df_2['train_rmse'], label='Train RMSE')
plt.plot(df_2.index, df_2['val_rmse'], label='Validation RMSE')
plt.ylabel('RMSE')
plt.legend()

# R^2 Plot
plt.subplot(3, 1, 2)
plt.plot(df_2.index, df_2['train_r2'], label='Train R^2')
plt.plot(df_2.index, df_2['val_r2'], label='Validation R^2')
plt.xlabel('Epochs')
plt.ylabel('R^2')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(df_2.index, df_2['lr'], label='learning rate')
plt.xlabel('Epochs')
plt.ylabel('lr')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 7))

# RMSE Plot
plt.subplot(3, 1, 1)
plt.plot(df_3.index, df_3['train_rmse'], label='Train RMSE')
plt.plot(df_3.index, df_3['val_rmse'], label='Validation RMSE')
plt.ylabel('RMSE')
plt.legend()

# R^2 Plot
plt.subplot(3, 1, 2)
plt.plot(df_3.index, df_3['train_r2'], label='Train R^2')
plt.plot(df_3.index, df_3['val_r2'], label='Validation R^2')
plt.xlabel('Epochs')
plt.ylabel('R^2')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(df_3.index, df_3['lr'], label='learning rate')
plt.xlabel('Epochs')
plt.ylabel('lr')
plt.legend()

plt.tight_layout()
plt.show()
```
